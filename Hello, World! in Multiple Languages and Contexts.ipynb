## Return 'Hello, World!' from a Function

```python
def hello_world() -> str:
    """Return a greeting message."""
    return "Hello, World!"
```

```python
if __name__ == "__main__":
    print(hello_world())
```

## Print 'Hello, World!' Using a Variable

```python
def main():
    message = "Hello, World!"
    print(message)
```

```python
if __name__ == "__main__":
    main()
```

## Print Multiple 'Hello, World!' Messages in a Loop

```python
# Print 'Hello, World!' messages multiple times using a loop
```

```python
def print_hello_messages(times: int) -> None:
    """Print 'Hello, World!' message a specified number of times."""
    for i in range(1, times + 1):
        print(f"{i}: Hello, World!")
```

```python
if __name__ == "__main__":
    print_hello_messages(5)
```
