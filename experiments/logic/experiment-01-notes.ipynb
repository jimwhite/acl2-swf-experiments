## Experiment: Three Ways Out of Check

**Location**: experiment-01-check-escape.lisp (also as `.ipynb`)

### What It Proves

The theorem `three-ways-out-of-check` formally states:

> If the king is in check and a move escapes check, then that move must be one of:
> 1. **King move** - the king itself moves to a safe square
> 2. **Capture** - a piece captures the attacking piece  
> 3. **Block** - a piece interposes between attacker and king

### Key Components

| Definition | Purpose |
|------------|---------|
| `square-p`, `piece-p` | Basic chess types |
| `position-p` | Game state (king + attacker + other pieces) |
| `in-check-p` | King is under attack |
| `is-king-move-p` | Move is by the king |
| `is-capture-p` | Move lands on attacker's square |
| `is-block-p` | Move is neither king-move nor capture |

### Theorems Proven

1. **`move-classification-exhaustive`** - Every move is king-move, capture, or block
2. **`block-excludes-king-move`** - Block and king-move are mutually exclusive
3. **`block-excludes-capture`** - Block and capture are mutually exclusive
4. **`three-ways-out-of-check`** - The main theorem
